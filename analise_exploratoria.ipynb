### 1\. Introdução

Este notebook apresenta uma análise exploratória com dados de sensores de equipamentos industriais. 
Os sensores industriais são utilizados para extrair dados em tempo real de determinado trecho de uma linha de produção. 
Esses dados permitem um controle eficaz sobre a linha, possibilitando identificar possíveis falhas, medição de peças e diversas outras aplicações.

### 2\. Análise de variáveis quantitativas

Com nossa base de dados numérica, vamos iniciar uma análise das variáveis quantitativas.

```python
# Importando Bibliotecas e arquivos

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv(r'data.csv', sep = ',', na_values = 'na', encoding = 'utf8')
pd.set_option('display.max_columns', None)
```

```python
# Amostra aleatória de 5 váriáveis no dataFrame

df.sample(5)
```

```python
# Renomeando coluna "Unamed para date"

df.rename(columns = {'Unnamed: 0': 'date'}, inplace = True)
```

```python
df.sample(5)
```

* Análise inicial (quantidade de linhas, média, min, max, desvio padrão e quartis)

```python
df.select_dtypes('number').describe().transpose()
```

### 3\. Agrupamento de colunas

* Separação das variáveis por categoria

```python
# Filtragem por categoria

df_temperaturas = df[['date', 'temperature_1', 'temperature_2', 'temperature_3', 'temperature_4', 'temperature_5', 'temperature_6', 'temperature_7']]
df_pressure = df[['date', 'pressure_1', 'pressure_2', 'pressure_3', 'pressure_4', 'pressure_5']]
df_speed = df[['date', 'speed_1', 'speed_2', 'speed_3']]
df_torque = df[['date', 'torque_1', 'torque_2', 'torque_3']]
```

```python
# Amostra aleatória de cada variável

display(df_temperaturas.sample(5))
display(df_pressure.sample(5))
display(df_speed.sample(5))
display(df_torque.sample(5))
```

### 4\. Visualização dos dados:

Atrávés da visualização das variáveis, vamos ter uma melhor dimensão da distribuição das frequências.

* Temperaturas

```python
axes = df_temperaturas.hist(figsize = (15, 9), bins = 50, color = 'mediumturquoise', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Pressão

```python
axes = df_pressure.hist(figsize = (12, 8), bins = 50, color = 'lavenderblush', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Velocidade

```python
axes = df_speed.hist(figsize = (10, 5), bins = 50, color = 'indigo', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Torque

```python
axes = df_torque.hist(figsize = (10, 7), bins = 50, color = 'lightblue', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

Através da visualização inicial, podemos ver algumas diferenças nas variáveis temperature_7 e torque_2 em comparação as demais de mesma classe.

### 4.1\. Correlação entre variáveis

Encontrando variáveis relacionadas e o quanto elas podem interferir entre si.

```python
df.corr()
```

```python
plt.figure(figsize = (13,10))
sns.heatmap(df.corr(), annot = True, fmt = '.2f', cmap = 'coolwarm', square = True)
plt.title('Mapa de Calor')
```
