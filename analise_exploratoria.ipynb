# 1 - Análise Exploratória de Dados (EDA)

```python
# Importando Bibliotecas e arquivos

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'data.csv', sep = ',', na_values = 'na', encoding = 'utf8')
pd.set_option('display.max_columns', None)
```

```python
df.sample(5)
```

```python
# Renomeando coluna "Unamed para date"
df.rename(columns = {'Unnamed: 0': 'date'}, inplace = True)
```

```python
df.sample(5)
```

### Analisando variáveis

```python
# Analisando variáveis

df.select_dtypes('number').describe().transpose()
```

### Agrupamento de colunas

```python
# Filtragem por categoria

df_temperaturas = df[['date', 'temperature_1', 'temperature_2', 'temperature_3', 'temperature_4', 'temperature_5', 'temperature_6', 'temperature_7']]
df_pressure = df[['date', 'pressure_1', 'pressure_2', 'pressure_3', 'pressure_4', 'pressure_5']]
df_speed = df[['date', 'speed_1', 'speed_2', 'speed_3']]
df_torque = df[['date', 'torque_1', 'torque_2', 'torque_3']]
```

```python
# Amostra aleatória de cada variável

display(df_temperaturas.sample(5))
display(df_pressure.sample(5))
display(df_speed.sample(5))
display(df_torque.sample(5))
```

## Visualização dos dados:

* Temperaturas

```python
axes = df_temperaturas.hist(figsize = (15, 9), bins = 50, color = 'skyblue', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Pressão

```python
axes = df_pressure.hist(figsize = (12, 8), bins = 50, color = 'skyblue', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Velocidade

```python
axes = df_speed.hist(figsize = (10, 5), bins = 50, color = 'skyblue', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

* Torque

```python
axes = df_torque.hist(figsize = (10, 7), bins = 50, color = 'skyblue', edgecolor = 'black')

for ax in axes.flatten():
    ax.set_ylabel('Frequency')

plt.show()
```

### Correlação entre variáveis

```python
df.corr()
```

```python
plt.figure(figsize = (13,10))
sns.heatmap(df.corr(), annot = True, fmt = '.2f', cmap = 'coolwarm', square = True)
plt.title('Mapa de Calor')
```
