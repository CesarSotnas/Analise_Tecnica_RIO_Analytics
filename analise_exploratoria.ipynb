# 1 - Análise Exploratória de Dados (EDA)

```python
# Importando Bibliotecas e arquivos

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'data.csv', sep = ',', na_values = 'na', encoding = 'utf8')
pd.set_option('display.max_columns', None)
```

```python
df.sample(5)
```

```python
df.rename(columns = {'Unnamed: 0': 'date'}, inplace = True)
```

```python
df.sample(10)
```

```python
# Analisando variáveis

df.select_dtypes('number').describe().transpose()
```

```python
df_temperaturas = df[['temperature_1', 'temperature_2', 'temperature_3', 'temperature_4', 'temperature_5', 'temperature_6', 'temperature_7']]
df_pressure = df[['pressure_1', 'pressure_2', 'pressure_3', 'pressure_4', 'pressure_5']]
df_speed = df[['speed_1', 'speed_2', 'speed_3']]
df_torque = df[['torque_1', 'torque_2', 'torque_3']]
```

```python
df_temperaturas.sample(10)
```

```python
df_temperaturas.hist(figsize = (15, 8), bins = 50, color = 'skyblue', edgecolor = 'black')
df_pressure.hist(figsize = (15, 8), bins = 30, color = 'skyblue', edgecolor = 'black')
df_torque.hist(figsize = (15, 8) , bins = 20, color = 'skyblue', edgecolor = 'black')
```

```python
plt.figure(figsize = (15,12))
sns.heatmap(df.corr(), annot = True, fmt = '.2f', cmap = 'coolwarm', square = True)
plt.title('Mapa de Calor')
```
